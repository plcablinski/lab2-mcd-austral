```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset_completo.csv",sep=";")
df.loc[df["estaVacunado_NA"]==True,"estaVacunado_No"] = True
df.loc[df["estaDesparacitado_NA"]==True,"estaDesparacitado_No"] = True
df.loc[df["estaEsterilizado_NA"]==True,"estaEsterilizado_No"] = True
df = df.drop(columns=["estaVacunado_NA","estaDesparacitado_NA","estaEsterilizado_NA"])
df.loc[df["Type"]==2,"es_raza_pura_Si"] = True
df["optimo"] = [True if (df["estaVacunado_Si"][i] and df["estaEsterilizado_Si"][i] and df["estaDesparacitado_Si"][i]) else False for i in range(len(df))]
df["calificacion"] = df[['estaVacunado_Si', 'estaDesparacitado_Si','estaEsterilizado_Si','estadoSalud_Saludable','SENTIMENT_POS']].sum(axis=1)
df['Representation'] = df['Representation'].apply(lambda x: eval(x))
lista_representation = [item for sublist in df['Representation'] for item in sublist]
lista_representation = list(set(lista_representation))
df["TOPIC_NAME_val"] = df["TOPIC_NAME"].str.split("_")
df["TOPIC_NAME_val"] = df["TOPIC_NAME_val"].apply(lambda x: int(x[0]))
df["Representation_val"] = df["Representation"].apply(lambda x: [lista_representation.index(i) for i in x])
listas_expandidas = pd.DataFrame(df['Representation_val'].to_list(), columns=['repre_1', 'repre_2', 'repre_3',
                                                                              'repre_4', 'repre_5', 'repre_6', 'repre_7', 'repre_8',
                                                                              'repre_9', 'repre_10'])
df["len_nouns"] = df["NOUNS"].apply(lambda x: len(x))
df['colores'] = df.apply(lambda row: [row['Color1'], row['Color2'], row['Color3']], axis=1)

for i in range(8):
    df[f"has_color{i}"] = [True if i in x["colores"] else False for index,x in df.iterrows()]

df.columns
```

```python
df.to_csv("dataset_2.csv",sep=";",index=False)
```
